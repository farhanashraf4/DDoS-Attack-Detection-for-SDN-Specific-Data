```python
import pandas as pd

# Load the CSV files into pandas DataFrames
df_tcp = pd.read_csv('BOUN_TCP_Anon.csv')
df_udp = pd.read_csv('BOUN_UDP_Anon.csv')

# Concatenate the DataFrames
concatenated_df = pd.concat([df_tcp, df_udp], ignore_index=True)

# Write the concatenated DataFrame to a new CSV file
concatenated_df.to_csv('concatenated_data.csv', index=False)
```

```python
import pandas as pd

def merge_column_data(file1, file2, output_file):
    # Read CSV files into pandas DataFrames
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)

    # Select common columns
    common_columns = ['Time', 'Source_ip', 'Destination_IP', 'Source_Port', 'Destination_Port', 'SYN', 'ACK', 'RST', 'Label']

    # Create a new DataFrame to store merged column data
    result_df = pd.DataFrame()

    # Concatenate column data for each DataFrame into a single column in the result DataFrame
    for column in common_columns:
        result_df[column] = pd.concat([df1[column], df2[column]])#,ignore_index=True)

    # Save the result DataFrame to a new CSV file
    result_df.to_csv(output_file)
    print(f"Merged column data saved to {output_file}")

if __name__ == "__main__":
    file1_path = "concatenated_data.csv"
    file2_path = "CICD.csv"
    output_file_path = "output_file.csv"

    merge_column_data(file1_path, file2_path, output_file_path)
```
